```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie_metadata = pd.read_csv(r"C:\Users\rudra\OneDrive\Desktop\VATMovieRecomm\CollaborativeMoviesDataset\movies_metadata.csv")
movie_metadata.head()
```

```python
movie_metadata = movie_metadata[['id','title', 'genres']]
movie_metadata
```

```python
movie_metadata.rename(columns = {'id':'movieId'}, inplace = True)
movie_metadata
```

```python
user_item_matrix = pd.read_csv(r"C:\Users\rudra\OneDrive\Desktop\VATMovieRecomm\CollaborativeMoviesDataset\user-item.csv")
user_item_matrix.head()
```

```python
# Define a KNN model on cosine similarity
from sklearn.neighbors import NearestNeighbors
cf_knn_model= NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)


# Fitting the model on our matrix
cf_knn_model.fit(user_item_matrix.transpose())
```

```python
#Selecting a random movie from the dataset and using that to generate a recommendation
n_recs = 10
movie_id=user_item_matrix.transpose().sample().index[0]
distances, indices = cf_knn_model.kneighbors(user_item_matrix[movie_id].values.reshape(1, -1), n_neighbors=n_recs)
movie_rec_ids = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
#print("original rec_ids")
#print(movie_rec_ids)
#print("descending rec_ids")
#print(sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1],reverse=True)[:0:-1])

# List to store recommendations
cf_recs = []
for i in movie_rec_ids:
    cf_recs.append({'Title':movie_metadata['title'][i[0]],'Distance':i[1]})

    
# Select top number of recommendations needed
df = pd.DataFrame(cf_recs, index = range(1,n_recs))

print("Recommendations if you liked ",movie_metadata.query(f"movieId == '{movie_id}'").title.iloc[0]," : ")
print(df)
```
