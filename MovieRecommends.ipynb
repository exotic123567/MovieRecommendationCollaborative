```python
#user_item_matrix = movie_data.pivot(index=['userId'], columns=['movieId'], values='rating').fillna(0)
user_item_matrix = movie_data.pivot_table(index='userId', columns='movieId', values='rating', aggfunc='mean').fillna(0)
user_item_matrix
```

```python
def movie_recommender_engine(movie_id, matrix, cf_model, n_recs):
    # Fit model on matrix
    #cf_knn_model.fit(matrix)
    
    # Extract input movie ID
    
    # Calculate neighbour distances
    distances, indices = cf_model.kneighbors(matrix[movie_id], n_neighbors=n_recs)
    movie_rec_ids = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
    
    # List to store recommendations
    cf_recs = []
    for i in movie_rec_ids:
        cf_recs.append({'Title':movie_metadata['title'][i[0]],'Distance':i[1]})
    
    # Select top number of recommendations needed
    df = pd.DataFrame(cf_recs, index = range(1,n_recs))
     
    return df
```

```python
user_item_matrix = pd.read_csv(r"C:\Users\rathi\OneDrive\Desktop\CollaborativeMoviesDataset\user-item.csv")
user_item_matrix.head()
```

```python
# Define a KNN model on cosine similarity
from sklearn.neighbors import NearestNeighbors
cf_knn_model= NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)


# Fitting the model on our matrix
cf_knn_model.fit(user_item_matrix.transpose())
```

```python
#Selecting a random movie from the dataset and using that to generate a recommendation
n_recs = 10
movie_id=user_item_matrix.transpose().sample().index[0]
distances, indices = cf_knn_model.kneighbors(user_item_matrix[movie_id].values.reshape(1, -1), n_neighbors=n_recs)
movie_rec_ids = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]

# List to store recommendations
cf_recs = []
for i in movie_rec_ids:
    cf_recs.append({'Title':movie_metadata['title'][i[0]],'Distance':i[1]})

# Select top number of recommendations needed
df = pd.DataFrame(cf_recs, index = range(1,n_recs))
```
